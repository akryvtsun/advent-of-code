```kotlin
val input = File("/Users/akryvtsun/Projects/advent-of-code/src/test/resources/year_2025/day_01_data.txt").readLines()
```

```kotlin
var position = 50
var count = 0
input.forEach { r ->
    val move = if (r[0] == 'L') -1 else 1
    val times = r.drop(1).toInt()
    repeat(times) {
        position += move
        if (position == -1) position = 99
        if (position == 100) position = 0
        if (position == 0) count++
    }
}
count
```
