```kotlin
fun isInvalid(n: Long): Boolean {
    val s = n.toString()
    for (i in 2..(s.length)) {
        if (s.length % i == 0) {
            val len = s.length / i
            val set = mutableSetOf<String>()
            set.addAll(s.chunked(len))
            if (set.size == 1) return true
        }
    }
    return false
}

input.split(",")
    .flatMap {
        val (from, to) = it.split("-")
        val res = mutableListOf<Long>()
        for (i in (from.toLong())..(to.toLong())) {
            if (isInvalid(i)) {
                res.add(i)
            }
        }
        res
    }.sum()
```
