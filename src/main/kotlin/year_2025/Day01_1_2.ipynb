```kotlin
val response = http.get("https://adventofcode.com/2025/day/1/input") {
    cookie("session", System.getenv("AOC_TOKEN"))
}
val input = response.bodyAsText().trimEnd()
```

```kotlin
input.lines()
    .runningFold(50) { position, cmd ->
        val direction = if (cmd[0] == 'L') -1 else 1
        val clicks = cmd.drop(1).toInt()
        (position + direction * clicks).mod(100)
    }.count { it == 0 }
```
